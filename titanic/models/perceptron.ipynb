```python
import pandas as pd
from expand import expand 
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from tensorflow.keras.layers.experimental.preprocessing import TextVectorization

from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.utils import to_categorical
from keras import optimizers
```

```python
temp = pd.read_csv("/home/jorge/Proyectos/datas/titanic/tables/data_cleaned/train.csv")
train = expand(temp)
```

```python
temp = train.join(pd.get_dummies(train[["Letter", "Embarked"]], dtype="int"))
```

```python
temp
```

## Data Processing

```python
del temp["Name"]
del temp["PassengerId"]
del temp["Letter"]
del temp["Ticket"]
del temp["Embarked"]
```

```python
temp["Sex"] = temp["Sex"].astype("int")
temp["Many"] = temp["Many"].astype("int")
temp["Survived"] = temp["Survived"].astype("int")
```

```python
temp["Age"] = pd.Series(preprocessing.scale(temp["Age"]))
temp["Fare"] = pd.Series(preprocessing.scale(temp["Fare"]))
temp["Number"] = pd.Series(preprocessing.scale(temp["Number"]))
```

```python
tempnona = temp.dropna()
```

## Spliting

```python
prop = 0.3/0.5
temp2, test = train_test_split(tempnona, test_size=0.5)
train, val = train_test_split(temp2, test_size=prop)
del temp2
```

```python
Input_dim = len(train.columns)-1
```

```python
Y_train = to_categorical(train["Survived"])
del train["Survived"]
X_train = train.values

Y_test = to_categorical(test["Survived"])
del test["Survived"]
X_test = test.values

Y_val = to_categorical(val["Survived"])
del val["Survived"]
X_val = val.values
validation = (X_val, Y_val)
```

## Training

```python
RN = Sequential()
RN.add(Dense(units=20, activation='sigmoid', input_dim=Input_dim))
RN.add(Dense(units=15, activation='sigmoid'))
RN.add(Dense(units=2, activation='softmax'))
```

```python
RN.summary()
```

```python
sgd = optimizers.SGD(lr=0.04, clipvalue=0.5, momentum=0.2)
```

```python
RN.compile(loss='categorical_crossentropy', optimizer=sgd, metrics=['accuracy'])
```

```python
%%time
history = RN.fit(X_train, Y_train, epochs=200, batch_size=1, validation_data=validation, verbose=0)
```

```python
RN.evaluate(X_test, Y_test, batch_size=1)
```
